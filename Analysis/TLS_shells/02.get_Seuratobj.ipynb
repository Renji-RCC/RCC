```r
library(Seurat)
library(ggplot2)
library(patchwork)
library(SeuratDisk)
library(SeuratObject)
library(dplyr)
library(reshape2)
library(openxlsx)
library(RColorBrewer)
library(gplots)
library(cowplot)
library(readxl)
library(dittoSeq)
library(ggpubr)
library(grid)
library(gridExtra)
library(circlize)
library(ComplexHeatmap)
```

```r
sample_list<-c('sample_tmp')
```

```r
for (sample in sample_list){
    output_path<-paste0('/data/work/TLS_shells/result/',sample)
    seu_path<-paste0('/data/work/cellbin/seu/',sample,'.Normalize.SCT.rds')
    seu<-readRDS(seu_path)
    expansion_data<-read.csv(paste0(output_path,'/',sample,'.expansion.only1200um.csv'),row.names=1)
    seu <- AddMetaData(object = seu,metadata = expansion_data,col.name = colnames(expansion_data))
    TLS_namelist <- unique(seu@meta.data[['TLS_name']])
    TLS_namelist <- TLS_namelist[TLS_namelist != "other"]
    for(TLSname in TLS_namelist){
        expansion_col <- paste0(TLSname,'_expansion')
        not_other <- seu@meta.data[[expansion_col]] != 'other'
        valid_tls <- seu@meta.data$TLS_name %in% c(TLSname, 'other')
        expansion_seu <- seu[, not_other & valid_tls]
        p <- SpatialDimPlot(expansion_seu, group.by = "TLS_name", pt.size.factor = 1.5, stroke = NA ) + 
            ggtitle(paste(TLSname)) +
            theme(legend.position = "right")
        ggsave(filename = paste0(output_path, '/seurat/', TLSname, '.only1200um_spatial.png'),plot = p,width = 8,height = 6,dpi = 300)
        saveRDS(expansion_seu,file=paste0(output_path,'/seurat/',TLSname,'.only1200um..expansion.rds'))
    }
}
```
