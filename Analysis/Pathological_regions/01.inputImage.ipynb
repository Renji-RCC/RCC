```r
cols <- list(
        'T'='#ec5e5d',
        'B'='#377EB8',
        'Plasma'='#fc98c7',
        'Myeloid'='#984EA3',
        'Mast'='#E3E114',
        'Endothelial'='#FC8D62',
        'Epithelial'='#17becf',
        'Myofibroblast'='#A65628'
    )
```

```r
filepath<-'/data/work/pathologicalRegion/celltype_info/sample_tmp_celltype.csv'
prefix <- '/data/work/pathologicalRegion/boundaryExpansion/1000x1000/sample_tmp'
```

```r
dat <- read.csv(filepath)
r1 <- range(dat$x)
r2 <- range(dat$y)

#### taking the maximum range for both X and Y
if(r1[2]-r1[1]>r2[2]-r2[1]){
  r2[2]=r2[1]+r1[2]-r1[1]
}else{
  r1[2]=r1[1]+r2[2]-r2[1]
}

ts <- unique(dat$main)
cols=cols[ts]
#### the color of each cell type
co <- cols[match(dat$main,ts)]


png(paste0(prefix,".fill.png"),width=1000,height=1000)
par(mar=c(0, 0, 0, 0), xaxs='i', yaxs='i')
plot(dat$x,dat$y,col=co,pch = 15,cex=0.02,xlim=r1,ylim=r2,axes=FALSE, frame.plot=F)
dev.off()
```
